```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=100)     # dimesions 5 by 3, to specify how many pixels it is pass second parameter dpi(pixels per inch)

# Keyword Argument Notation
# plt.plot(x, y, label='2x', color='r', linewidth=2, marker='.', linestyle='--', markersize=10 , markeredgecolor='blue')

# Use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2
# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)      # [0.  0.5 1.  1.5 2.  2.5 3.  3.5 4. ]
print(x2)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--', label='x^2')


# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict= {'fontname': 'Comic Sans MS', 'fontsize' : 20})

# X and Y labels
plt.xlabel('X Axis', fontdict= {'fontname': 'Comic Sans MS', 'fontsize' : 10})
plt.ylabel('Y Axis', fontdict= {'fontname': 'Comic Sans MS', 'fontsize' : 10})


# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# to save the graph (saves it in the same directory)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

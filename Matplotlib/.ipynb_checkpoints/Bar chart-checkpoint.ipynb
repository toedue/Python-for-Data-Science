### Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as ps 
import numpy as np
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

plt.bar(labels, values)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

bars = plt.bar(labels, values)
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```

# If Statement

if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>

```python
age = 14

if age>=18:
    print("You're an adult!")
elif age >= 13:
    print("You're a teenager")
else:
    print("You're a kid")
```

# For loop 

for <variable> in <iterable>:
    <code>

```python
countries
```

```python
for country in countries:
    print(country)
```

```python
for i, country in enumerate(countries):
    print(i, country)
```

```python
my_data = {'name': 'Abudy', 'age': 22, 'height': 1.8}
my_data 
```

```python
for key, value in my_data.items():
    print(key, value)
```

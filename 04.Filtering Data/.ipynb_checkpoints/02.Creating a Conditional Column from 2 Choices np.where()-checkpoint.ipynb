# Creating a Conditional Column from 2 Choices: np.where()

```python
# we import numpy bexause where method belongs to numpy
import numpy as np
```

```python
df_laptops['Price_euros']
```

```python
df_laptops['Price_euros'] > 2000
```

```python
# create an array based on price tiers
np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive') 
# the first arg is the condition, 
# the second arg is the value we wanna set if the condition is true
# the third arg is the value we wanna set if the condition is false
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive')
```

```python
df_laptops.head()
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```

# Exercise

```python
# create an array based in condition(screen size > 15)
np.where(df_laptops['Inches']>15, 'Big', 'Small')
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
df_laptops.head()
```

```python
print(df_laptops.columns)
```

```python
df_laptops.drop(['Screen_sizes'], axis=1, inplace=True)
```

```python
# count the values in screen_size column
df_laptops.value_counts("Screen_size")
```

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding="latin-1")
```

```python
df_laptops
```

## isin(): Single Filtering

```python
# select Apple or HP Laptops
df_laptops['Company'].isin(['Apple', 'HP']) # inside a list
```

```python
# filter dataframes
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])]
```

```python
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])].value_counts("Company")
```

## isin(): Multiple Filtering

```python
# find notebooks or ultabooks from Apple or HP
filter1 = df_laptops['Company'].isin(['Apple', 'HP'])
filter2 = df_laptops['TypeName'].isin(['Notebook', 'Ultrabook'])
```

```python
filter1 & filter2
```

```python
# filter dataframe (link the filters with a logical operator)
df_laptops[filter1 & filter2]
```

```python
df_laptops[filter1 & filter2]["Company"].value_counts()
```

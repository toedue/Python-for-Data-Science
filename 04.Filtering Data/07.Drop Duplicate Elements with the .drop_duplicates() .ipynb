```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding="latin-1")
```

```python
df_laptops.head(3)
```

# drop_duplicates()
drop_duplicates() removes rows that have duplicate values.

```python
# remove duplicate elements inside the copany colums
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# verify
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']].value_counts("Company")
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# cheapest: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'
# First drop duplicates with inplace=True
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
# Then select the columns
result = df_laptops[['Company', 'Price_euros']]
result  # Display the result
```

```python
# Argument

# inplace: drop duplicates in place or to return a copy.
# ignore_index=True: the resulting axis will be labeled 0, 1, …
```

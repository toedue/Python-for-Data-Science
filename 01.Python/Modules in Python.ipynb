## 1️⃣ `import os`

```python
import os
```

* This imports the **OS module** in Python.
* The **OS module** lets Python interact with your **Operating System** (Windows, Linux, Mac).

You can use it to:

* check folders
* create folders
* delete files
* see the current directory
* etc.

---

## 2️⃣ `os.getcwd()`

```python
os.getcwd()
```

`getcwd` means:

**get current working directory**

👉 It shows the **folder where your program is currently working**.

---

## 3️⃣ Your Output

```
'C:\\Users\\Hp\\Desktop\\My_course'
```

This means your **current folder is:**

```
C:\Users\Hp\Desktop\My_course
```

### `os.listdir()` 

`os.listdir()` is a function from the **OS module** that shows **all files and folders inside a directory**.

---

## 1️⃣ Syntax

```python
import os
os.listdir(path)
```

* **path** → the folder you want to check
* If you **don’t give a path**, it shows files from the **current working directory**.

---

## 2️⃣ Example (Current Folder)

```python
import os

os.listdir()
```

### Example Output

```python
['data.csv', 'notes.txt', 'project.py', 'images']
```

os.listdir("C:\\Users\\Hp\\Desktop")
```

This will show **everything inside the Desktop folder**.

Example output:

```python
['My_course', 'python_notes.pdf', 'music', 'todo.txt']
```

---

`os.listdir()` → **returns a list of all files and folders in a directory**

### `os.makedirs("New Folder")` 

`os.makedirs()` is used to **create a new folder (directory)**.

---

## 1️⃣ Syntax

```python
import os
os.makedirs("folder_name")
```

* **`folder_name`** → name of the folder you want to create.

---

## 2️⃣ Example

```python
import os

os.makedirs("New Folder")
```

### What happens?

Python creates a **new folder** called:

```
New Folder
```

inside your **current working directory**.

Earlier your directory was:

```
C:\Users\Hp\Desktop\My_course
```

So the new folder will be created here:

```
C:\Users\Hp\Desktop\My_course\New Folder
```

---



## 3️⃣ Creating Multiple Folders at Once

`makedirs` can create **nested folders** automatically.

```python
import os

os.makedirs("project/data/files")
```

This creates:

```
project
   └── data
        └── files
```

Even if the folders **don’t exist yet**.

---

## 4️⃣  Important Note ⚠️

If the folder **already exists**, Python will give an error.

Example:

```python
os.makedirs("New Folder")
```

Error:

```
FileExistsError
```

To avoid this:

```python
os.makedirs("New Folder", exist_ok=True)
```

Now it **won’t crash if the folder already exists**.

## 1️⃣ `os.remove()`

**Definition:**
Deletes a **file** from the system.

**Syntax**

```python
import os
os.remove("file_name")
```

**Example**

```python
import os

os.remove("data.txt")
# Output: data.txt file deleted
```

**Important**

* Works only for **files**
* Gives error if used on **folders**

---

## 2️⃣ `os.rmdir()`

**Definition:**
Deletes an **empty folder (directory)**.

**Syntax**

```python
import os
os.rmdir("folder_name")
```

**Example**

```python
import os

os.rmdir("New Folder")
# Output: New Folder deleted
```

**Important**

* Works only if the folder is **empty**
* If the folder has files → error occurs

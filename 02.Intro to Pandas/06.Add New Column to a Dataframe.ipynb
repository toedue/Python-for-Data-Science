# Adding a new column with an array
The dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0, 1000) # gives us range that starts with first argument and to the last argument
```

```python
language_score
```

```python
# verify the length of the array because length of the array must be the same as the length of the dataframe
len(language_score)
```

```python
# adding a new column to dataframe with an array
df_exams['language score'] = language_score
df_exams 
```

```python
# create random integer numbers between 1 and 100 (method-> random.randint)
int_language_score = np.random.randint(1, 100, size=1000) 
# first arg -> lowest value of the random numbers (inclusive)
# second arg -> the highest value in the random numbers (exclusive)
# third arg -> is the size 

int_language_score
```

```python
print(len(int_language_score))

# min value inclusive and high value exclusive
print(min(int_language_score))
print(max(int_language_score))
```

```python
# adding a new column to dataframe with an array
df_exams['language score'] = int_language_score
df_exams
```

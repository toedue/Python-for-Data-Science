```python
import pandas as pd
```

```python
# reading and showing dataframe
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams
```

# Create Index

```python
import numpy as np
import random
```

```python
# creating non repetitive values for the index
new_index = np.arange(0, 1000)
new_index
```

```python
# shuffling the indexes using random library
random.shuffle(new_index)
new_index
```

# create a new column with new index (set new_index variable to a new column of the orginal dataframe)
df_exams['new_index'] = new_index
df_exams

# Set Index

```python
# set new_index column as of the dataframe by the method (set_index)
df_exams.set_index('new_index', inplace=True) #used inplace because the index method only creates a copy
df_exams
```

# Sort Index

```python
# sort dataframe by new index
df_exams.sort_index()
```

```python
# sort descending by index
df_exams.sort_index(ascending=False)
```

```python
# sort descending by index
df_exams.sort_index(ascending=False, inplace=True)
```
